```python
import pandas as pd
import numpy as np
```

```python
def montar_df(periodo, valor, banco, tipo):
  df = pd.DataFrame()
  df['periodo'] = periodo
  df['valor'] = valor
  df['banco'] = banco
  df['tipo'] = tipo
  return df
```

```python
 #Itau derivativos
periodo = [202212, 202303, 202306, 202309, 202312]
valor = [78.341, 88.687, 72.773, 68.730, 56.383]
df = montar_df(periodo, valor, 'itau', 'atv')
```

```python
#Itau derivativos
valor = [78.512, 84.582, 68.027, 64.087, 53.495]
df_aux = montar_df(periodo, valor, 'itau', 'pas')
```

```python
df = pd.concat([df, df_aux], axis=0)
```

```python
# Bradesco
valores = [16.591, 28.207, 23.513, 22.870, 15.492]
df_aux = montar_df(periodo, valor, 'bradesco', 'atv')

df = pd.concat([df, df_aux], axis=0)
valores = [11.536, 24.043, 21.856, 19.936, 13.835]
df_aux = montar_df(periodo, valor, 'bradesco', 'pas')
df = pd.concat([df, df_aux], axis=0)
```
